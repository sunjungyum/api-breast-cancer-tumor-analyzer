```python
# Importing necessary libraries
import urllib
import requests
import time
import random
from flask import Flask, request
import json
import pandas as pd### K Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
import numpy as np
import os

# Setting up current directory:
CURRENT_DIRECTORY = os.getcwd()
```

### K Nearest Neighbors

```python
# Importing data from a csv file
dataset = pd.read_csv(CURRENT_DIRECTORY + '/breast_cancer_dataset.csv')

# Adding an Id tag to the dataframe
dataset['Id'] = range(569)
### K Nearest Neighbors
# Ways to visualize the data:
rows, cols = dataset.shape

# Replace outliers with median value:
outliers = np.where(dataset['bare_nuclei'] <= -100000)[0]
for outlier in outliers:
    dataset['bare_nuclei'][outlier] = dataset['bare_nuclei'].median()
    
# Check how many of each class (benign vs. malignant) we have
dataset.groupby('class').size()

# Splitting up the labels and the values for each species:
feature_columns = ['clump_thickness', 'uniformity_of_cell_size', 'uniformity_of_cell_shape', 
                   'marginal_adhesion', 'single_epithelial_cell_size', 'bare_nuclei', 'bland_chromatin',
                   'normal_nucleoli', 'mitosis']

X = dataset[feature_columns].values
Y = dataset['class'].values
```

```python
# Data visualization
plt.figure(figsize=(15,10))
parallel_coordinates(dataset.drop("Id", axis=1), "class", color = ['red','green'])
plt.title('Parallel Coordinates Plot', fontsize=20, fontweight='bold')
plt.xlabel('Attributes', fontsize=15)
plt.ylabel('Attributes values', fontsize=15)
plt.legend(loc=1, prop={'size': 15}, frameon=True, shadow=True, facecolor="white", edgecolor="black")
plt.show()
```

### Training the model:

```python
# Splitting into training and test datasets:
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.5, random_state = 0)

# Creating the learning model
knn_classifier = KNeighborsClassifier(n_neighbors=10)

# Fitting the model with the training data
knn_classifier.fit(X_train, Y_train)
```

### Test the model

```python
# Making predictions with the test data (This line is also where we would potentially classify new data)
Y_pred = knn_classifier.predict(X_test)
print(Y_pred)

# Finding Accuracy:
accuracy = accuracy_score(Y_test, Y_pred)*100
print('Accuracy of model: ' + str(round(accuracy, 2)) + ' %.')
cm = confusion_matrix(Y_test, Y_pred)
cm
```

### Testing out different k values

```python
# creating list of cv scores
cv_scores = []
k_list = [i for i in range(1, 50, 2)]

# perform 10-fold cross validation
for k in range(1, 50, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, Y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())

# Displaying results visually
plt.figure()
plt.figure(figsize=(15,10))
plt.title('The optimal number of neighbors', fontsize=20, fontweight='bold')
plt.xlabel('Number of Neighbors K', fontsize=15)
plt.ylabel('Accuracy', fontsize=15)
plt.plot(k_list, cv_scores)

plt.show()
```

## Making a breast cancer tumor predictor API

```python
# Set up Flask App
app = Flask(__name__)

@app.route("/", methods = ['GET'])
def classify():
    # get values for each component, return error message if not a float
    try:
        values = [[float(request.args.get(component)) for component in ['clump_thickness', 'uniformity_of_cell_size', 'uniformity_of_cell_shape', 
                   'marginal_adhesion', 'single_epithelial_cell_size', 'bare_nuclei', 'bland_chromatin',
                   'normal_nucleoli', 'mitosis']]]
    except TypeError:
        return "An error occured\nUsage: 127.0.0.1:5000?clump_thickness=NUM&uniformity_of_cell_size=NUM&uniformity_of_cell_shape=NUM&marginal_adhesion=NUM&single_epithelial_cell_size=NUM&bare_nuclei=NUM&bland_chromatin=NUM&normal_nucleoli=NUM&mitosis=NUM"
    
    # Otherwise, return the prediction.
    prediction = knn_classifier.predict(values)[0]
    if prediction == 2:
        return "Tumor is benign"
    else:
        return "Tumor is malignant"

# Run the app.
app.run()

# try 127.0.0.1:5000?clump_thickness=4&uniformity_of_cell_size=3&uniformity_of_cell_shape=1&marginal_adhesion=1&single_epithelial_cell_size=2&bare_nuclei=10&bland_chromatin=3&normal_nucleoli=1&mitosis=1
    
```
